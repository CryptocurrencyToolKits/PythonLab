## 测试与数据库相关的库函数

### 1.  从数据库获得tick数据并显示

```python
# 从数据库获得价差数据
from vtFunction import *
from ctaBase import *
import pandas as pd
import numpy as np
fields = ['datetime','askPrice1','bidPrice1','turnover','volume']
symbol = u'rb1705'
# 从数据库获得数据
tickers1 = loadHistoryData(TICK_DB_NAME,symbol,'20161011 20:59:00','20161011 21:02:00',fields)
buy = tickers1['bidPrice1']-tickers1['bidPrice1'].mean()
sell = tickers1['askPrice1']-tickers1['bidPrice1'].mean()
# 计算tick成交均价
turnover = tickers1['turnover']-tickers1['turnover'].shift(1)
turnover = turnover.apply(lambda x:max(x,0))
volume = tickers1['volume']-tickers1['volume'].shift(1)
volume = volume.apply(lambda x:max(x,1))
aprice = turnover/volume/10
aprice = aprice - tickers1['bidPrice1'].mean()
aprice = aprice[aprice > -2]
spread0 = aprice - buy - 10
spread1 = aprice - sell - 10
datas = pd.DataFrame({'buy':buy,'sell':sell,'aprice':aprice,'spread0':spread0,'spread1':spread1})
# 画图
import matplotlib.pyplot as plt
plt.close()
datas=datas.dropna(axis=0, how='any')
datas
datas.plot(kind='line')
plt.show()
```
