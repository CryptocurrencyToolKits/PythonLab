## 黑色系主成分套利分析

>黑色系合约 : 螺纹，铁矿石，热卷，焦炭

>分析时间   : 最近3年

>原理和目标 : 利用主成分分析，用于找出协整残差项，利用残差项的均值回归特性，在残差大幅度偏离均值时进场，在残差回归均值时出场

### 1. 导入需要用到的库和函数

>sklearn PCA库没有给出具体的主成分向量，所以需要自己编写部分函数

```python
from sklearn.decomposition import PCA
from datetime import datetime,timedelta
from vtFunction import *
from ctaBase import *
import matplotlib.pyplot as plt
import numpy as np

class handleRet(object):

    def __init__(self,portfolio,date,period = u''):
        ####the input of the portfolio
        self.ptf = [index for index in portfolio]
        self.period = period
        
        ####the output of  the portfolio
        self.currDate = date
        self.e = np.ones((len(portfolio),1))
        self.exsRet = self.exsRet(self.period)
        self.retMat = self.retMat()
        self.retCov = self.retCov()
        
    def exsRet(self,period):
        if period == u'':
            if self.period == u'':
                period = 100
            else:
                period = self.period
                
        fields = ['datetime','open','close']        
        endDate = self.currDate.strftime("%Y%m%d")
        beginDate = self.currDate - timedelta(days=period)
        beginDate = beginDate.strftime("%Y%m%d")
        ticker = [index for index in self.ptf]
        
        pdData = {index:loadHistoryData(DAILY_DB_NAME,index,beginDate,endDate,fields) for index in ticker}
        exsRet = {}
        for index in ticker:
            # print pdData[pdData.ticker==index]['closePrice']
            pdData[index]['preclose'] = pdData[index]['close'].shift(1)
            pdData[index] = pdData[index].dropna(axis=0, how='any')
            exsRet[index] = (pdData[index]['close'] - pdData[index]['preclose'])/pdData[index]['close'].mean()
        return exsRet
    
    def retMat(self):
        ticker = sorted(self.exsRet.keys(),key = lambda x:x)
        # print ticker
        tmpList = []
        for index in ticker:
            tmpList.append(self.exsRet[index].tolist())
        return np.mat(tmpList)
        pass
    
    def retCov(self):
        try:
            return np.mat(np.cov(self.retMat))
        except Exception,e:
            pass
            # print 'The retCov of is unavailable, please change your period!'

def meanX(dataX):
    return np.mean(dataX,axis=0)#axis=0表示按照列来求均值，如果输入list,则axis=1

def pca(XMat, k):
    """
    参数：
        - XMat：传入的是一个numpy的矩阵格式，行表示样本数，列表示特征    
        - k：表示取前k个特征值对应的特征向量
    返回值：
        - finalData：参数一指的是返回的低维矩阵，对应于输入参数二
        - reconData：参数二对应的是移动坐标轴后的矩阵
    """
    average = meanX(XMat) 
    m, n = np.shape(XMat)
    data_adjust = []
    avgs = np.tile(average, (m, 1))
    data_adjust = XMat - avgs
    covX = np.cov(data_adjust.T)   #计算协方差矩阵
    featValue, featVec=  np.linalg.eig(covX)  #求解协方差矩阵的特征值和特征向量
    index = np.argsort(-featValue) #按照featValue进行从大到小排序
    finalData = []
    if k > n:
        print "k must lower than feature number"
        return
    else:
        #注意特征向量时列向量，而numpy的二维矩阵(数组)a[m][n]中，a[1]表示第1行值
        selectVec = np.matrix(featVec.T[index[:k]]) #所以这里需要进行转置
        finalData = data_adjust * selectVec.T 
        reconData = (finalData * selectVec) + average  
    return index[:k], selectVec
```

### 2. 分析3阶主成分的解释能力和相关性
>从结果可以看到，3阶主成分已经具有足够的解释能力

```python
####合约池、测试日期
universe = ['RB0000','J0000','I0000','HC0000']
currDate = datetime.now() - timedelta(days=300)

####收益率矩阵获取
retData = handleRet(universe,currDate,period = 600)
retmat = retData.retMat.T

####对收益率矩阵做3阶PCA拟合
pca0 = PCA(n_components = 3)
pcaFitData = pca0.fit_transform(retmat)

####PCA分析结果
print '3阶主成分方差解释比例：\n' + str(np.sum(pca0.explained_variance_ratio_))
print '\n'
print '3阶主成分方差权重分配：\n' + str(pca0.explained_variance_ratio_)
# for index in np.shape(pca.explained_variance_ratio_)[1]
print '\n'
print '第一主成分与第二主成分之间的相关系数为: ' + str(np.corrcoef(pcaFitData[:,0],pcaFitData[:,1])[0,-1])
print '\n'
print '3阶主成分特征向量图'
plt.close()
for index in range(np.shape(pcaFitData)[1]):
    plt.plot(pcaFitData[:,index])
plt.show()    
```

### 3. 使用3阶主成分表示螺纹主力合约每日收益
>3阶主成分很好地解释了螺纹主力的每日收益

```python
from sklearn import linear_model

clf = linear_model.LinearRegression()

clf.fit(pcaFitData,retmat[:,0])
corr = clf.coef_
fits = np.mat(pcaFitData)*corr.T

plt.close()
plt.plot(fits.tolist(),'r')
plt.plot(retmat[:,0].tolist(),'g')
plt.legend(['PCA regression result','RB0000'])
plt.show()
```

### 4. 焦炭合约资金曲线减去3阶主成分后的残差图

```python
# 使用的主成分数量
num = 3

# 计算回报曲线
capList = []
for i in xrange(4):    
    cap = np.cumsum(retmat[:,i])+1
    cap = cap.tolist()[0]
    capList.append(cap)
capMat = np.mat(capList)
featVal, featVec = pca(retmat,3)

print featVec

# 计算各个主成分大小
pcaList = []
for i in xrange(3):   
    pca0 = featVec[i,:]*capMat
    pca0 = pca0.tolist()[0]
    pcaList.append(pca0)
pcaMat = np.mat(pcaList)
    
# 计算主成分在各个合约价格上的表示   
fitList = []
for i in xrange(4): 
    fit0 = featVec[0:num,i].T*pcaMat[0:num]
    fit0 = fit0.tolist()[0]
    fitList.append(fit0)
    
# 计算残差
errList = []
for i in xrange(4): 
    err = np.array(capList[i])-np.array(fitList[i])
    errList.append(err)
index = 0
factor =[0]*4
for i in xrange(num):
    for j in xrange(4):
        factor[j] +=  featVec[i,index]*featVec[i,j]
print list([-factor[0],(1-factor[1]),-factor[2],-factor[3]])

err0 = -factor[0]*np.array(capList[0])+(1-factor[1])*np.array(capList[1])-factor[2]*np.array(capList[2])-factor[3]*np.array(capList[3])
plt.close()
#plt.plot(capList[0],'g')
#plt.plot(fitList[0],'r')
plt.plot(errList[1],'g')
plt.plot(err0,'r')
plt.show()
```
